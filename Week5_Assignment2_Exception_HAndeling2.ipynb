Q1. Explain why we have to use the Exception class while creating a Custom Exception.

In Python Custom Exceptions are defined by creating a Custom Exception class that inherits the "Exception" class of one of its subclass.
Using Exception class as the base class for our Custom Exception class makes our codebase consistent as in Python Exceptions follow a Hierarchy and by using Exception class as our base class, our Custom Exception also becomes a part of this Hierarchy. 
Due to this Hierarchy it becomes easier for us to write the "except" block like.
Also using Exception class as our base class facilitates us with various inbuilt functionalities like displaying custom message for our exception.

Q2. Write a python program to print Python Exception Hierarchy.

```python
def Print_Exception_Hierarchy(exception_class,depth=0):
    space="  " * depth
    print(f"{space}{exception_class.__name__}")
    
    for subclass in exception_class.__subclasses__():
        Print_Exception_Hierarchy(subclass,depth+1)
        
Print_Exception_Hierarchy(Exception)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

There are three errors defined in ArithmeticError class
1. ZeroDivisionError
2. OverflowError
3. FloatingPointError

1. ZeroDivisionError: This error is thrown by the python interpreter when we try to divide some numaric value by zero.

```python
try:
    print(10/0)
except ZeroDivisionError as e:
    print(e)
```

2. OverflowError: When we are performing some operations on a numaric data and the result is too large to be accomadated by that data type then OverflowError is thrown by the Python interpreter.

```python
import math
try:
    print(math.exp(10000))
except OverflowError as e:
    print(e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError class serves as a base class for errors related to the lookup on iterables such as List, tuples, dictionary.
KeyError:- This error is raised when we try to access value corresponding to a non exixting key in a dictionary.
IndexError:- This error is raised when we try to access value at an index greater than the maximum index of the sequence data types.

```python
#KeyError

Details={"First_Name":"Anmol", "Last_Name":"Anand", "Email":"Anmol65438@gmail.com", "Course":"DataScienceMasters2.0"}

try:
    print(Details["Phone"])

except KeyError as e:
    print(e)
```

```python
#IndexError

lst=[1,2,3,4,5,6]

try:
    print(lst[9])
    
except IndexError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is raised when there is some issue with the module that we are trying to import. The issue can be any thing related to the module like module with given name do not exist, some kind of syntax error or any other error in the in the module, issue with some dependencies of the module. 
ModuleNotFoundError class is a subslass of the ImportError Class. ModuleNotFoundError is raised only when we are trying to import a module that do not exist.
ImportError Class is more generic and donot provide much information about the reason behid the error, Where as ModuleNotFoundError Class is specific only to one situation that is when module donot exist. 

```python
#ImportError
try:
    import Random_module
except ImportError as e:
    print(e)
```

```python
#ModuleNotFoundError
try:
    import Random_module
except ModuleNotFoundError as e:
    print(e)
```

Q6. List down some best practices for exception handling in python.

Following are some best practices for exception handeling in Python:
1. Always try to catch specific errors in the 'except' block instead of using 'Exception" class.
2. Avoid using multiple except blocks. But if you are using multiple except blocks to catch specific errors and if you are using "Exception" class as a default case than that except block must be placed at the bottom after all except blocks.
3. Always try to log the error instead of just printing it on console bacause the console will be cleared as soon as the program or the system restarts but the log is maintained in permanent file.
4. Try to make your custom ecxeption classes wherever required to handle situations that may be syntactically correct but will produce undesired result.
5. Always try to log meaningful statements with complete information which will make the debugging easier in case of errors.
6. Always cleanup all the resources utilised during the program execution.
