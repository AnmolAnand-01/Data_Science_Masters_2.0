**Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.**

```python
import seaborn as sns
import plotly.graph_objects as go

titanic_data=sns.load_dataset("titanic")
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic_data.age, y=titanic_data.fare, mode="markers")) 
fig.update_layout(
    title="Scatter Plot: Age VS Fare",
    xaxis=dict(title='Age'),
    yaxis=dict(title='Fare')
)

fig.show()
```

**Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.**

```python
import plotly.express as px
import seaborn as sns

# Load the 'tips' dataset from seaborn
tips_data = sns.load_dataset('tips')

# Create facetted subplots using Plotly Express
fig = px.box(tips_data, x='day', y='total_bill', color='sex', facet_row='time', facet_col='smoker',
             title='Box Plot of Total Bill by Day, Gender, Time, and Smoker')
fig.update_xaxes(title='Day')
fig.update_yaxes(title='Total Bill')

# Show the plot
fig.show()
```

```python
# fig = px.box(tips_data, x='day', y='tip', color='sex', title='Box Plot of Tip by Day and Gender')
fig = px.box(tips_data, x='day', y='tip', color='sex', facet_row='time', facet_col='smoker',
             title='Box Plot of Tip by Day, Gender, Time, and Smoker')
fig.update_xaxes(title='Day')
fig.update_yaxes(title='Tip')

# Show the plot
fig.show()
```

**Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.**

```python
import seaborn as sns
import plotly.express as px

tips_data=sns.load_dataset("tips")

fig=px.histogram(tips_data, x="sex", y="total_bill", pattern_shape="smoker", color="day", 
                 title="Histogram of Total Bill by Sex, Day and Gender")
fig.update_yaxes(title='Total Bill')
fig.update_xaxes(title='Sex')
fig.show()
```

**Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.**

**Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.**

```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns

# Load the Iris dataset
iris= sns.load_dataset("iris")

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(
    iris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species',
    title='Scatter Matrix Plot of Iris Dataset',
    height=800, width=800
)

# Update marker properties
fig.update_traces(marker=dict(size=5))

# Show the plot
fig.show()
```

**Q5. What is Distplot? Using Plotly express, plot a distplot.**

"Distplot" is commonly used to plot the distribution of univariate (single variable) datasets. It combined a histogram with a kernel density estimation (KDE) plot and, optionally, rug plots by default.

Plotly Express did not have a direct equivalent function named "distplot".

```python
import plotly.express as px
import seaborn as sns

# Load sample data (for example purposes)
tips_data = sns.load_dataset('tips')

# Create a distribution plot using Plotly Express - Histogram
fig = px.histogram(tips_data, x='total_bill', marginal='rug', title='Distribution Plot of Total Bill')
fig.update_xaxes(title='Total Bill')
fig.update_yaxes(title='Frequency')

# Show the plot
fig.show()
```
