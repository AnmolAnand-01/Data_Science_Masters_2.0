**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

Matplotlib is a widely used python library for creating static, intractive and publication-quality visualisations. It offers various tols for generating various types of plots and graphs, making it a versatile tool for data visualisation.

Below are the five plots that can be ploted using Pyplot module of Matplotlib library in python.

1. Line Plot: It shows data points connected by straight line segments, often used to display trends over time or continuous data.

2. Scatter Plots: Represents indivisual data points as markers, helpful for visualisaing the relationship between two variables.

3. Histogram: Displays the distribution of single numerical variable by dividing the data into bins and showing the frequency of values within each bin.

4. Bar plot: Uses bars to represent categorial data. It compares different categories or groups by their value.

5. Pie Charts: Illustrates the portion or percentage distribution of categorial data as slice of circular pie.

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**
 
**import numpy as np**
 
**np.random.seed(3)**
 
**x = 3 + np.random.normal(0, 2, 50)**
 
**y = 3 + np.random.normal(0, 2, len(x))**

**Note: Also add title, xlabel, and ylabel to the plot.**

Scatter plor is a type of data visualisation, that displays individual data point as marker on a two-dimensional cartesian coordinate system. It represents the relationship between two variables, typically plotting one variable on x-axis and the other variable on y-axis.

```python
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel(" Data on X-axis ")
plt.ylabel(" Data on Y-axis ")
plt.title(" Scatter plot for X vs Y ")
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:**
 
**import numpy as np**
 
**For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])**

**For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])**

**For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])**

**For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])**

"subplot()" function is used to create a grid of subplots within a single figure. It allows to display multiple plots or graphs in different sections or different cells of grid, enabling the visualistion of several datasets or comparisions in a single figure. 

```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.subplot(2,2,1)
plt.plot(x1,y1,'r')
plt.xlabel("Value of X1")
plt.ylabel("Value of Y1")
plt.title("Plot number 1")

plt.subplot(2,2,2)
plt.plot(x2,y2,'g')
plt.xlabel("Value of X2")
plt.ylabel("Value of Y2")
plt.title("Plot number 2")

plt.subplot(2,2,3)
plt.plot(x3,y3,'b')
plt.xlabel("Value of X3")
plt.ylabel("Value of Y3")
plt.title("Plot number 3")

plt.subplot(2,2,4)
plt.plot(x4,y4,'y')
plt.xlabel("Value of X4")
plt.ylabel("Value of Y4")
plt.title("Plot number 4")

plt.tight_layout()
plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

**import numpy as np**

**company = np.array(["Apple", "Microsoft", "Google", "AMD"])**

**profit = np.array([3000, 8000, 1000, 10000])**

A bar plot is a graphical representation used to display categorial data. It presents data in rectangular bars where the length of each bar corresponds to the value it represents.

Bar plots can be used for many purpose including:

1. Comparing frequencies or count of different categories.

2. Showing the distribution of categorial data.

3. Showing changes or trends in categorial data over time or across different conditions.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))

#Bar Plot
plt.subplot(1,2,1)
plt.bar(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Bar plot for Company VS Profit")

#Horizontal Bar Plot
plt.subplot(1,2,2)
plt.barh(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Horizontal Bar plot for Company VS Profit")

plt.tight_layout()
plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

**box1 = np.random.normal(100, 10, 200)**

**box2 = np.random.normal(90, 20, 200)**

A box plot, also known as box-and-wishker plot, is a statical visualisation method, used to display the distribution, central tendency and dsplay of a dataset.

Box plot is used because of varaious reasons, some of them are following:- 

1. It helps in identifying the potetial outliers or extreme values within a dataset, making them valuable in identifying anomalies and extreme observations.

2. They give the snapsot of key statical measures sucha as the median quartiles and range in a visually understandable format.

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

data_lst=[box1,box2]
plt.boxplot(data_lst,labels=['Data-1', 'Data-2'])

plt.title('Box Plot Example')
plt.xlabel('Data Sets')
plt.ylabel('Values')

plt.show()
```
