**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Seaborn is a powerful python data visualisation library, built on top of Matplotlib. It offers various types of plots to visualise different aspects of data. 

Below are the five commomly used plots with theig typical use cases:-

1. Histogram: Displays the distribution of single numerical variable by dividing the data into bins and showing the frequency of values within each bin.

2. Scatter Plots: Represents indivisual data points as markers, helpful for visualisaing the relationship between two variables.

3. Bar plot: Uses bars to represent categorial data. It compares different categories or groups by their value.

4. Box plots: Show the distribution of data based on summary statistics like median, quartiles and outliers.

5. Heatmaps: Helps in visualise matrix-like data using colors to represent values.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

**Note: timepoint, signal, event, and region are columns in the fmri dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri=sns.load_dataset("fmri")
plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title('Signal Over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.**

**Note: pclass, age, and fare are columns in the titanic dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
titanic=sns.load_dataset("titanic")
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Boxplot for pclass VS age")
plt.xlabel("pclass")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Boxplot for pclass VS fare")
plt.xlabel("pclass")
plt.ylabel("Fare")
plt.show()
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
diamonds=sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut")
plt.title("Histogram for frequency of the price range of diamonds")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

**Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**

```python
import seaborn as sns

iris=sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
```

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')

plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
